```python
import pickle
import matplotlib.pyplot as plt
import glob
import os
import numpy as np

root = '../annotation/'
files = glob.glob(root+'2017*')

label_dict= pickle.load(open(os.path.join(root, 'label_dict.pkl'), 'r'))
print label_dict
```

```python

l = []
for f in files:
    label = pickle.load(open(os.path.join(f,'annotations.pkl'), 'r'))
    plt.figure('Labels', figsize=(15, 6))
    plt.plot(label[:,0])
    plt.plot(label[:,1])
    plt.show()
    
    label = np.sum(label, axis=1)
    print f
    
    l.append( np.sum(label==0) / float(label.shape[0]))
    print "percentage of background: ", np.sum(label==0) / float(label.shape[0])

print "Average percentage: ", np.mean(np.asarray(l))

#label = pickle.load(open(os.path.join(f,'annotations.pkl'), 'r'))

```

```python
from scipy.stats.stats import pearsonr   

distribution_all = np.zeros(17)

for f in files:
    label = pickle.load(open(os.path.join(f,'annotations.pkl'), 'r'))
    
    for i in np.nditer(label):
        if not i==0:
            distribution_all[i-1] += 1

print "All labels distribution"
plt.figure('Label distribution')
plt.plot(distribution_all)
plt.show()

for f in files:
    label = pickle.load(open(os.path.join(f,'annotations.pkl'), 'r'))
    
    distribution = np.zeros(17)
    for i in np.nditer(label):
        if not i==0:
            distribution[i-1] += 1
            
    plt.figure('Label distribution')
    plt.plot(distribution)
    plt.show()
    
    print f
    print pearsonr(distribution, distribution_all)
    
print "Example: 201704151140"
```
